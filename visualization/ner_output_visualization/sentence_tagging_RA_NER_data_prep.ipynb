```python
#separate the individual sentences to get just the indices of sentence endings
indexes_sentence_end = []
for i in range(0, len(text)):
    if (text.iloc[i] == ".") or (text.iloc[i] == "!") or (text.iloc[i] == "?"):
        if len(text.iloc[i]) == 1:
            indexes_sentence_end.append(i)
        
        
    
```

```python
#getting just the subset of data to visualize
indexes_sentence_end_tr = indexes_sentence_end[:171]
```

```python
list_text = text.to_list()
```

```python
lists = []
for i in range(0, len(indexes_sentence_end_tr)-4, 1):
    lists.append(list_text[indexes_sentence_end_tr[i]+1:indexes_sentence_end_tr[i+1]+1])
    
    
```

```python
lists_ner = []
for i in range(0, len(indexes_sentence_end_tr)-4, 1):
    lists_ner.append(predicted_labels[indexes_sentence_end_tr[i]+1:indexes_sentence_end_tr[i+1]+1])
    
    
```

```python
#preparing data for visualization
list_dict = []
for var in range(0, len(lists)):
    dict_content_ner  = {}
    dict_content_ner["content"] = lists[var]
    dict_content_ner["path"] = "data-enron-subset/inbox/2."
    dict_content_ner["ner"] = lists_ner[var]
    list_dict.append(dict_content_ner)
    
```

```python
dict_sent = {}
dict_sent["sentences"] = list_dict
```
